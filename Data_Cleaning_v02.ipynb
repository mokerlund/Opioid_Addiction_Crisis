# Data analysis using completely cleaned data ONLY

```python
import pandas as pd

number_of_deaths = pd.read_csv("clean data/2018_deaths_clean.csv") # good
hotline_calls = pd.read_csv("Clean Data/county_hotline_calls_2018.csv") # good
drug_boxes = pd.read_csv("Clean Data/drug_boxes_by_county.csv") # good
pa_population = pd.read_csv("clean data/PA_population_clean.csv") # good
od_deaths = pd.read_csv("clean data/OD_Deaths_Cleaned_More.csv") # good
emergency_od_visits = pd.read_csv("Clean Data/further_cleaned_od_ed_visits_2018.csv") # Unclear Column names, needs more cleaning
```

```python
merged_1 = pd.merge(number_of_deaths, hotline_calls, on='County', how='outer')
merged_2 = pd.merge(merged_1, drug_boxes, on='County', how='outer')
merged_3 = pd.merge(merged_2, pa_population, on='County', how='outer')
merged_4 = pd.merge(merged_3, od_deaths, on='County', how='outer')
opioid_pa_df = pd.merge(merged_4, emergency_od_visits, on='County', how='outer')

opioid_pa_df.head()
# Opioid_pa_df = pd.merge(number_of_deaths, hotline_calls, drug_boxes, pa_population, od_deaths, emergency_od_visits, on='County', how='Outer')
```

```python
opioid_pa_df = opioid_pa_df.sort_values(by='County')
print(opioid_pa_df.columns)


opioid_pa_df=opioid_pa_df.rename(columns={"2018":"hotlinecall_2018"})
opioid_pa_df.columns
```

## Further cleaned by Elle

```python
# calculate death ratios, drop columns, reformatting

opioid_pa_df2=opioid_pa_df.drop(['Jan18', 'Feb18', 'Mar18', 'Apr18','May18', 'Jun18', 'Jul18', 'Aug18', 'Sep18', 'Oct18', 'Nov18', 'Dec18'],axis=1)
opioid_pa_df2["%_of_any_od_death_over_all_deaths"]=(opioid_pa_df2["Any Drug Overdose Deaths"]/opioid_pa_df2["death_count"]*100).map("{:.2f}".format)
opioid_pa_df2["%_of_op_od_death_over_all_deaths"]=(opioid_pa_df2["Opioid Overdose Deaths"]/opioid_pa_df2["death_count"]*100).map("{:.2f}".format)
opioid_pa_df2["Annual Any Overdose ED Rate by Visit"] = opioid_pa_df2["Annual Any Overdose ED Rate by Visit"].map("{:.2f}".format)
opioid_pa_df2["Annual Opioid Overdose ED Rate by Visit"] = opioid_pa_df2["Annual Opioid Overdose ED Rate by Visit"].map("{:.2f}".format)
opioid_pa_df2["Annual Any Overdose ED Rate by Population"] = opioid_pa_df2["Annual Any Overdose ED Rate by Population"].map("{:.2f}".format)
opioid_pa_df2["Annual Opioid Overdose ED Rate by Population"] = opioid_pa_df2["Annual Opioid Overdose ED Rate by Population"].map("{:.2f}".format)

# change column order

opioid_pa_df2.columns

col_new = ['County','2019_Population','death_count','Any Drug Overdose Deaths','Opioid Overdose Deaths',\
           '%_of_any_od_death_over_all_deaths','%_of_op_od_death_over_all_deaths',\
           'Annual Any Overdose ED Rate by Visit','Annual Opioid Overdose ED Rate by Visit',\
           'Annual Any Overdose ED Rate by Population','Annual Opioid Overdose ED Rate by Population',\
           'hotlinecall_2018', 'Number Boxes']
           
opioid_pa_df2=opioid_pa_df2[col_new]
```
